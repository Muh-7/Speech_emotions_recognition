```python
import os
from IPython.display import Audio
import librosa
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from tensorflow.keras.layers import Dense, Dropout, LSTM, TimeDistributed, Flatten, BatchNormalization
from tensorflow.keras.models import Sequential
from keras.layers import Conv1D
from tensorflow.keras.optimizers import Adam
from keras.callbacks import ReduceLROnPlateau, EarlyStopping
from tensorflow.keras.models import load_model
import pandas as pd
from sklearn.metrics import confusion_matrix
import seaborn as sns
```

```python
path = "/content/drive/MyDrive/emotions/"

def load_data(path):
    f_emotions =[]
    f_pathes = []
    folders = os.listdir(path)
    for folder in folders:
        files = os.listdir(path+folder)
        for file in files:
            step = file.split('.')[0]
            step = step.split('-')[2]
            f_emotions.append(int(step))
            f_pathes.append(path + folder + os.sep+ file)
    return [f_emotions,f_pathes]

def get_emotion (number):
    info ={
        1 : 'nuetral',
        2 : 'calm',
        3 : 'happy',
        4 : 'sad',
        5 : 'angry',
        6 : 'fearfull',
        7 : 'disgust',
        8 : 'surprised'
        }
    return info[number]

emotions, pathes = load_data(path)
```

```python
def read_audio(path):
    data, sample_rate = librosa.load(path, duration=2.4, offset =0.6)
    return data, sample_rate


def draw_wave(path,data,sr):
    plt.figure(figsize = (10,4))
    #data, sr = librosa.load(path)
    plt.title('aiudio wave ::'+ path, size = 14)
    librosa.display.waveshow(data, sr = sr)
    plt.show()

def draw_spectogram(path,data,sr):
    plt.figure(figsize = (10,4))
    #data, sr = librosa.load(path)
    X = librosa.stft(data)
    Xdb = librosa.amplitude_to_db(np.abs(X))
    plt.title('aiudio wave ::'+ path, size = 14)
    librosa.display.specshow(Xdb, sr = sr, x_axis = 'time', y_axis = 'log')
    plt.show()
```

```python
def add_noise(data, sr):
    noise = 0.035*np.random.uniform()*np.amax(data)
    data += noise * np.random.normal(size=data.shape[0])
    return data, sr

def shift(data, sr):
    shift_range = int(np.random.uniform(low = -5, high = 5)*1000)
    shifted = np.roll(data, shift_range)
    return shifted, sr

def pitch(data, sr, factor= 0.7):#try(0.7,10,100)
    pitched = librosa.effects.pitch_shift(y=data, sr=sr, n_steps= factor)
    return pitched, sr

def stretch(data, sr, rate=0.95):#try (0.5,1,1.55,2)
    stretched = librosa.effects.time_stretch(y=data, rate=rate)
    return stretched, sr

#Feature Extraction MFFCs:
def feature_extraction(data, sr):
    mfcc = librosa.feature.mfcc(y=data, sr =sr)
    return mfcc

def processing_audio(data, sr, option):
    func = random.choice(option)
    if func == 'standard':
        processed = data
    else :
        processed, _ = func(data, sr)

    return processed
```

```python
def get_features(path):
    data, sample_rate = read_audio(path)
    funcs = ['standard', add_noise, pitch]

    features = []

    func1_data = processing_audio(data, sample_rate, funcs)
    func2_data = processing_audio(func1_data, sample_rate, funcs)


    feature= feature_extraction(func2_data,sample_rate)

    if feature.shape == (20, 104):
        features.append(feature)

    func1_data = processing_audio(data, sample_rate, funcs)
    func2_data = processing_audio(func1_data, sample_rate, funcs)


    feature= feature_extraction(func2_data,sample_rate)

    if feature.shape == (20, 104):
        features.append(feature)

    func1_data = processing_audio(data, sample_rate, funcs)
    func2_data = processing_audio(func1_data, sample_rate, funcs)


    feature= feature_extraction(func2_data,sample_rate)

    if feature.shape == (20, 104):
        features.append(feature)


    return features


def display(number):
    data, sample_rate = read_audio(pathes[number])
    mfcc_feature = feature_extraction(data, sample_rate)
    print(mfcc_feature)
    print(get_emotion(emotions[number]))
    #data, sample_rate = add_noise(data,sample_rate)
    data, sample_rate = pitch(data,sample_rate)
    data, sample_rate = stretch(data,sample_rate)
    draw_wave(pathes[number],data, sample_rate)

    draw_spectogram(pathes[number],data, sample_rate)

    return data, sample_rate


d,sr = display(70)
Audio(data= d, rate= sr)
```

```python
X = []
Y = []

for indx in range(len(pathes)):
    value = get_features(pathes[indx])
    if value != []:
        for item in value:
            X.append(item)
            Y.append(np.eye(8)[emotions[indx] - 1]) #0 --> 7
```

```python
for indx in range(len(pathes)):
    print(get_emotion(emotions[indx]))
```

```python
encoder = OneHotEncoder()
encoder.fit_transform(np.array([1,2,3,4,5,6,7,8]).reshape(-1,1)).toarray()
```

```python
x_train, x_test, y_train, y_test = train_test_split(np.array(X), np.array(Y), test_size=0.2, random_state = 11, shuffle=True)
print('trian x shape', x_train.shape)
print('test x shape', x_test.shape)
print('trian y shape', y_train.shape)
print('test y shape', y_test.shape)
```

```python
trainx = np.expand_dims(x_train, axis = 3)
trainx = np.expand_dims(trainx, axis = 3)
trainx = np.swapaxes(trainx,1,2)
print('trainx : ',trainx.shape)

testx = np.expand_dims(x_test, axis = 3)
testx = np.expand_dims(testx, axis = 3)
testx = np.swapaxes(testx,1,2)
print('testx  : ',testx.shape)

inputShape = trainx.shape[1:]
inputShape
```

```python
def createModel(inputShape):
    model = Sequential()
    model.add(TimeDistributed(Conv1D(32,3,padding = 'same', activation = 'relu'), input_shape = inputShape))
    model.add(TimeDistributed(BatchNormalization()))
    model.add(TimeDistributed(Flatten()))

    model.add(LSTM(64))
    model.add(Dropout(0.2))

    model.add(Dense(units=64,activation= 'relu'))
    model.add(Dropout(0.2))
    model.add(Dense(units=64,activation= 'relu'))

    model.add(Dense(units=8,activation= 'softmax'))

    return model


```

```python
model = createModel(inputShape)
model.summary()
```

```python
opt = Adam(learning_rate = 0.01)
model.compile(optimizer = opt, loss = 'categorical_crossentropy', metrics = ['accuracy'])

reduce = ReduceLROnPlateau(monitor = 'val_loss', factor = 0.6, verbose = 1, patience = 5, min_lr = 1e-8)
stop = EarlyStopping(monitor = 'val_loss', patience = 7)

hist = model.fit(trainx, y_train, batch_size = 140, epochs = 80, validation_data = (testx, y_test), callbacks = [reduce, stop]) #to draw it after the training

```

```python
print('The accuracy:', model.evaluate(testx, y_test)[1]*100, '%')
train_loss = hist.history['loss']
test_loss = hist.history['val_loss']
train_accuracy = hist.history['accuracy']
test_accuracy = hist.history['val_accuracy']

epochs = [value for value in range(69)]
print(len(train_loss))
print(len(epochs))

fig, ax = plt.subplots(1,2)
fig.set_size_inches(15,6)

ax[0].plot(epochs,train_loss, label = 'Training_loss')
ax[0].plot(epochs,test_loss, label = 'Testing_loss')
ax[0].set_title('Test and Training Loss')
ax[0].legend()
ax[0].set_xlabel('Epochs')
ax[0].set_ylabel('Loss')


ax[1].plot(epochs,train_accuracy, label = 'Training_Accuracy')
ax[1].plot(epochs,test_accuracy, label = 'Testing_Accuracy')
ax[1].set_title('Test and Training Accuracy')
ax[1].legend()
ax[1].set_xlabel('Epochs')
ax[1].set_ylabel('Accuracy')

model.save(r'/content/drive/MyDrive/emotions/emotion_model.h5')
```

```python
import librosa
import numpy as np

def read_audio(path):
    data, sample_rate = librosa.load(path, duration=2.4, offset=0.6)
    return data, sample_rate
```

```python
def feature_extraction(data, sr):
    mfcc = librosa.feature.mfcc(y=data, sr=sr, n_mfcc=20)
    return mfcc
```

```python
def process_audio(path):
    data, sample_rate = read_audio(path)

    # Extract MFCC features
    mfcc_features = feature_extraction(data, sample_rate)

    # Transpose to match the shape (timesteps, features)
    mfcc_features = np.transpose(mfcc_features)

    # Ensure the shape is (1, 20, 160, 1) for prediction
    mfcc_features = np.expand_dims(mfcc_features, axis=-1)
    mfcc_features = np.expand_dims(mfcc_features, axis=0)

    return mfcc_features
```

```python
def predict_emotion(model, path):
    # Process the audio file to get features
    processed_audio = process_audio(path)

    # Predict using the model
    predictions = model.predict(processed_audio)

    # Get the emotion with the highest probability
    predicted_emotion = get_emotion(np.argmax(predictions))

    return predicted_emotion
```
